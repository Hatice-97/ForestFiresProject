```python
#53.Bar chart
DMC = [23,85, 72, 43, 52]
DC= [42, 35, 21, 16, 9]
width =0.3
plt.bar(np.arange(len(DMC)), DMC, width=width)
plt.bar(np.arange(len(DC))+ width, DC, width=width)
plt.show()
```

```python
#53.Bar chart
DMC = [0.1, 17.5, 40, 48, 52, 69, 88] #bar chart daha iyii
DC = [2, 8, 70, 1.5, 25, 12, 28]
index = ['mon', 'tue', 'wed',
         'thu', 'fri', 'sat', 'sun']
df = pd.DataFrame({'DMC': DMC,
                   'DC': DC}, index=index)
ax = df.plot.bar(rot=0)
```

```python
#54.Pie chart
day = ['mon','tue','wed','thu','fri','sat','sun']
pay = [112,58,40,60,40,110,71]
colors = ['yellow','green','red','blue']
plt.pie(pay,labels=day,colors=colors)
plt.axis('equal')
```

```python
#55. Heat map
# create tabular correlation matrix 
corr = df.corr()
_, ax = plt.subplots(figsize=(13,10)) 
# graph correlation matrix
_ = sns.heatmap(corr, ax=ax,
xticklabels=corr.columns.values,
yticklabels=corr.columns.values,
cmap='coolwarm')
```

```python
#56.Seaborn Kutuphanesi kullanarak Stripplot grafigi
#import seaborn as sns
p = sns.stripplot(data=df,
                  y='DC',
                  x='temp',
                  jitter=0.25,
                  dodge=False)
```

```python
#58.Swarmplot grafigini ciziniz
#import seaborn as sns
p = sns.swarmplot(data=df,
                  y='DC',
                  x='temp',
                  dodge=False)
```
